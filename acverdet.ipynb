```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("acverdetfinal.csv")
```

```python
mag = df[list(df)[0]]
phi = df[list(df)[1]]
dphi = df[list(df)[2]] 
x = np.linspace(0, .025, 100)
```

```python
from matplotlib import rc
plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')
phifinal = 10.0*phi
dphifinal = 10.0*dphi
```

```python
import scipy.stats as sp
slope, intercept, r_value, p_value, std_err = sp.linregress(mag, phifinal)
slope
intercept
y = slope*x + intercept
```

```python
plt.errorbar(mag, phifinal, fmt = 'bv', 
             yerr=dphifinal, label = r"$\Delta \Phi$" )
plt.plot(x, y, label = r"Linear Fit $y = 2B +.0013$")
plt.xlabel(r"\textbf{Magnetic Field} \textit{(Tesla)}")
plt.ylabel(r"$\Delta \Phi$ \textit{(Radians)}")
plt.title(r"Change in Plane of Polarization ($\Delta \Phi$) v. Magnetic Field (AC)")
plt.legend(loc="lower right")
plt.show()
```
