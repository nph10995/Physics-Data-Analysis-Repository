```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import math
```

```python
df = pd.read_csv("maluslaw2.csv")
```

```python
angle = df[list(df)[0]]
```

```python
intensitydf = df[list(df)[1]]
```

```python
clean_angle = [] 
dirty_index = []
for i in range(len(angle)):
    if angle[i]%5==0:
        clean_angle.append(angle[i])
    else:
        dirty_index.append(i)
print(dirty_index)
print(clean_angle)
```

```python
intensity = []
for i in intensitydf:
    intensity.append(i/1000)
print(type(intensity))
```

```python
clean_intensity=[]
for i in range(len(intensity)):
    if i != dirty_index[0]:
        clean_intensity.append(intensity[i])

print(len(clean_angle))
print(len(clean_intensity))
m = max(clean_intensity)
```

```python
radians = []
for i in clean_angle:
    r = i*((np.pi)/(180))
    radians.append(r)

theory = m*(np.cos(radians)**2)
```

```python
plt.plot(radians, clean_intensity, 'r-', label = "Data Collected")
plt.plot(radians, theory)
plt.show()
```
