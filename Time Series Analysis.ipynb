```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
GU = pd.read_csv("GBPUSD1d.csv")
list(GU)
```

```python
biddf = GU['1.55358']
askdf = GU['1.55371']
```

```python
bid = list(biddf)
ask = list(askdf)
len(bid)
```

```python
price = []
for i in range(len(bid)):
    price.append((bid[i]+ask[i])/2)
```

```python
index = np.linspace(0,62010, 62011)
```

```python
price_sqrt =[]
for i in price: 
    price_sqrt.append(np.sqrt(i))
```

```python
price_cubert = []
for i in price:
    price_cubert.append(i**(1/3))
```

```python
bid_distance = []
for i in range(len(bid)):
    bid_distance.append(price[i]-bid[i])
```

```python
plt.plot(index,price)
```

```python
plt.plot(index, bid_distance)
```
